**Structural range (below Tg):** The semi-crystalline thermoplastic as well as the amorphous thermoplastic polymer is stiff and brittle in the glass region, given the rotational and displacement barriers hindering any kind of mobility. The chains are 'frozen' in place, which results in a low free volume. The temperature in this case is well below the glass transition temperature Tg. The storage modulus E' is the highest in this region for both thermoplastic polymers, with the semi-crystalline generally being higher compared to its amorphous counterpart. The structural range ends near Tg for both, however the semi-crystalline polymer partially retains some mechanical support from crystalline regions, since those require much higher energy to disbond. For both tanδ is very low, indicating that there is little to no molecular motion lagging behind applied deformation.

**In the vicinity of Tg:** With an increasing temperature, the localized mobility starts increasing coinciding with the increase of free volume. A higher mobility results in the reduction of stiffness, hence E' drops. Both polymers showcase a peak in tanδ, suggesting a high energy dissipation, meaning that molecular motion lags strongly behind applied deformation. The semi-crystalline polymer retains some stiffness due to crystals resulting in a less pronounced drop in E' (slower softening) compared to the amorphous polymer.

**Rubbery plateau:** In this region, the free volume becomes more significant coinciding with a higher chain mobility. E' is low but steady in both thermoplastic polymer types, but higher in semi-crystalline thermoplastic polymers still due to the rigid crystalline regions that provide mechanical support, keeping the material stiffer than the amorphous counterpart. Tanδ of the semi-crystalline thermoplastic polymer steadily increases whereas tanδ of their amorphous counterpart is decreasing.

**Melting temperature (semi-crystalline only):** At this point the crystals melt, resulting in a sharp drop in E', since the chains become fully mobile coinciding with a sharp increase in free volume. Tanδ showcases a second peak, suggestion another spike in energy dissipation upon rupture of crystal bonds. At the same high temperature, the amorphous thermoplastic polymers are already soft, having a relatively low E' and tanδ.

**Processing range**: Both thermoplastic polymers can be processed. Both have a very high free volume and an extremely low E' with the characteristic of the semi-crystalline thermoplastic polymer behaving like their amorphous counterpart, since the crystals melted away and don't provide any more mechanical reinforcement.

#### The longitudinal coefficient of linear thermal expansion for a composite ply can be calculated/estimated as follows: 
$$\alpha_{l} = \frac{(1-\nu_{f})E_{r}\alpha_{r}+ \nu_{f}E_{lf} \alpha_{lf}}{(1-\nu_{f})E_{r} +\nu_{f}E_{lf}}$$
#### The transverse coefficient of linear thermal expansion for a composite ply can be calculated/estimated as follows:
$$\alpha_{t} = (1-\nu_{f})\alpha_{r}+ \nu_{f}\alpha_{tf} +(1-\nu_{f})\alpha_{r}\nu_{r} +\nu_{12f}\alpha_{lf}\nu_{f} -\nu_{12}\alpha_{l}$$

#### Where $\nu_{f}$ represents fibre volume content, $E_{r}$ [GPa] Young's modulus of the matrix, $\nu_{r}$ the Poisson ratio of the matrix, $\alpha_{r}$ [1/°C] coefficient of linear thermal expansion of the matrix, $\alpha_{lf}$ [1/°C] and $\alpha_{tf}$ [1/°C] thermal expansion coefficients (CLTE) of fibres in the longitudinal and transverse direction, $E_{lf}$ [GPa] the longitudinal Young's modulus of the fibres, $\nu_{12}$ the in-plane Poisson ratio of the lamina and $\nu_{12f}$ Poisson’s ratio of the fibres.

#### The thermal strain (in longitudinal and transverse direction) can be calculated as follows: 
#### $$\epsilon = \int_{0}^{t} \alpha(t)\Delta Tdt$$

#### Question 3.1: Given the provided equations, material properties, consolidation cycle, and the listed assumptions, please calculate:
#### - the composite material's linear coefficient of thermal expansion (in longitudinal and transverse direction).
#### - the expected strain after reaching room temperature in both length and width direction of the panel
#### - the percentage of shrinkage in length and width direction of the panel


#### Note - You may use Kulkarni R, Ochoa O. Transverse and Longitudinal CTE Measurements of Carbon Fibers and their Impact on Interfacial Residual Stresses in Composites. Journal of Composite Materials. 2006;40(8):733-754. doi:10.1177/0021998305055545 as reference for material data.

```python
import dataclasses

@dataclasses.dataclass
class PolymerProperties:
    name: str
    youngs_modulus: float  # GPa
    poisson_ratio: float
    clte: float  # 1/°C

@dataclasses.dataclass
class FiberProperties:
    name: str
    youngs_modulus: float  # GPa
    poisson_ratio: float
    clte_longitudinal: float  # 1/°C
    clte_transverse: float  # 1/°C

polycarbonate = PolymerProperties(
    name="Polycarbonate",
    youngs_modulus=1.4e9,
    poisson_ratio=0.37,
    clte=70.2e-6
)

polyamide = PolymerProperties(
    name="Polyamide 6",
    youngs_modulus=3e9,
    poisson_ratio=0.39,
    clte=85e-6
)

# Fiber properties
t_1000 = FiberProperties(
    name="T1000 UD",
    youngs_modulus=290e9,  # GPa
    poisson_ratio=0.2,
    clte_longitudinal=5.0e-6,  # 1/°C
    clte_transverse=4.9e-6  # 1/°C
)

vf = 0.55  # fibre volume fraction


# \alpha_{l} = \frac{(1-\nu_{f})E_{r}\alpha_{r}+ \nu_{f}E_{lf} \alpha_{lf}}{(1-\nu_{f})E_{r} +\nu_{f}E_{lf}}
alpha_l = ((1-t_1000.poisson_ratio) * polycarbonate.youngs_modulus * polycarbonate.clte +
            vf * t_1000.youngs_modulus * t_1000.clte_longitudinal) / \
           ((1 - vf) * polycarbonate.youngs_modulus + vf * t_1000.youngs_modulus)

alpha_t = ((1 - vf) * polycarbonate.clte +
            vf * t_1000.clte_transverse +
            (1 - vf) * polycarbonate.clte * polycarbonate.poisson_ratio +
            vf * t_1000.clte_longitudinal * t_1000.poisson_ratio -
            (1 - t_1000.poisson_ratio) * alpha_l)
print(f"Longitudinal CTE (Polycarbonate): {alpha_l:.2e} 1/°C")
print(f"Transverse CTE (Polycarbonate): {alpha_t:.2e} 1/°C")

# Thermal strain calculation
delta_T = 200 - 20  # °C
epsilon_l = alpha_l * delta_T
epsilon_t = alpha_t * delta_T
print(f"Longitudinal strain: {epsilon_l:.2e}")
print(f"Transverse strain: {epsilon_t:.2e}")
# Percentage shrinkage
shrinkage_l = epsilon_l * 100  # %
shrinkage_t = epsilon_t * 100  # %
print(f"Longitudinal shrinkage: {shrinkage_l:.2f}%")
print(f"Transverse shrinkage: {shrinkage_t:.2f}%")
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Path to image (relative to the notebook). Adjust if needed.
img_path = "slice_0130.jpeg"

# Load and convert to grayscale
img = Image.open(img_path).convert("L")  # "L" = 8-bit grayscale

# Convert to numpy array and flatten
arr = np.array(img)
flat = arr.ravel()

# Count pixels per brightness value 0..255
counts = np.bincount(flat, minlength=256)  # length 256 ensures full 0-255 range
brightness = np.arange(256)

# Plotting (use log y-axis; skip zero-count bins so log scale works)
pos = counts > 0
plt.figure(figsize=(8,4))
plt.plot(brightness[pos], counts[pos], color="tab:blue", lw=1)
plt.fill_between(brightness[pos], counts[pos], color="tab:blue", alpha=0.2)
plt.xlabel("Brightness (0 = black, 255 = white)")
plt.ylabel("Number of pixels")
plt.title(f"Pixel count vs brightness for '{img_path}'\nTotal pixels = {flat.size:,}")
plt.grid(alpha=0.3, which='both')
plt.xlim(0,255)
plt.yscale('log')
plt.tight_layout()
plt.show()
```
